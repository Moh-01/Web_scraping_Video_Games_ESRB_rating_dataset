# Problem identification:

Many parents are giving their children any game to play without thinking about what's the content of that game or what's the target age for it. and this is affecting the children specifically, because many games are not made for kids, either because of sexual content, drugs, and/or violence. ESRB works to empower parents to make informed decisions about the video games they purchase for their children and allow them to be played in their homes. The ESRB doesn’t aim to prohibit certain games from being purchased and played but rather hopes to provide concise and impartial information to parents and other consumers.

- ESRB rating: An assigned rating category that suggests the age-appropriateness of that particular game

---------

# Goal:

Predict the ESRB rating for games based on the content that the game rated on.

---------

# Data dictionary:

- This data contains the name for `1895 games` with `34 of ESRB rating content` as features for each game.<br><br>
- A single datapoint is represented as a single number form `0-3` for `Console` and a `binary vector` for the `features of ESRB content`.

# Extract rating list

- I'll scrape the rating features becuase there are different category for each rate. ex(Blood, mid blood, simulated blood)
- I think 7 pages is enough to get all of the rates.

```python
rating_features = []
driver = webdriver.Chrome(executable_path='chromedriver/chromedriver')

for pg in range(1,8):
    
    games_list   = []
    
    #Change the url number
    page = f'https://www.esrb.org/search/?searchKeyword=&platform=PlayStation%204&rating=E%2CE10%2B%2CT%2CM%2CAO&descriptor=All%20Content&pg={pg}&searchType=All'

    #Get the page
    driver.get(page)
    
    #Wait for 5s
    sleep(5) 
    
    #Downlaod the .html page
    html = driver.page_source 

    #Read the page
    webpage = BeautifulSoup(html,'html.parser')
    
    #Get the products
    games_list = (webpage.find_all('div', attrs={'class':'game'}))

    
    for i in games_list:
        rating_features.append(i.find('div', attrs={'class':'content'}).find('table').find_all('td')[1].string.split(','))
    
    print(f'Page ===> [{pg}] ===> Extracted successfully!')
```

```python
rating_features
```

```python
#Clean the output and remove the duplicated with spaces
rating_feat = sorted(list({x.replace(' ','') for i in rating_features for x in i}))
rating_feat
```

# Automated crawling using Selenium - Train data

```python
#load Chrome driver
driver = webdriver.Chrome(executable_path='chromedriver/chromedriver')

#------------------------------------------------Initializing--------------------------------------------------#
#Webscraping - itterate through 81 pages
for pg in range(1,201):
    
    games_list = []
    title   = []
    console = []
    rating  = []
    esrb_rating = []
    
#--------------------------------------------------Reading-----------------------------------------------------#
    
    #Change the url number
    page = f'https://www.esrb.org/search/?searchKeyword=&platform=PlayStation%204&rating=E%2CE10%2B%2CT%2CM%2CAO&descriptor=All%20Content&pg={pg}&searchType=All'

    #Get the page
    driver.get(page)
    
    #Wait for 5s
    sleep(5) 
    
    #Downlaod the .html page
    html = driver.page_source 

    #Read the page
    webpage = BeautifulSoup(html,'html.parser')
    
    #Get the products
    games_list = (webpage.find_all('div', attrs={'class':'game'}))
    
#-------------------------------------------------Scraping-----------------------------------------------------#

    #Itterate through gpu_list
    for i in games_list:
        
        #Create the vectore of rating 25 - 0's 
        rating_vector = [0]*31
        
        #-----------------------------Add title-----------------------------------#
        
        try:
            title.append(i.find('div', attrs={'class':'heading'}).find('a').string)
        except:
            title.append(np.nan)
            
        #-----------------------------Add console---------------------------------#
        #ps4 = 0, Xbox = 1, both = 2, 3 = other
        try:
            #Fet the consels for the games
            con = i.find('div', attrs={'class':'heading'}).find('div',class_='platforms').string.replace(' ', '').split(',')
            
            #Check if it's ps4 or Xbox_one or both
            if 'PlayStation4' in con and 'XboxOne' in con:
                console.append(2)
                
            elif 'PlayStation4' in con:
                console.append(0)
            
            elif 'XboxOne' in con:
                console.append(1)
            else:
                console.append(3)
                
        except:
            console.append(np.nan) 
            
        #-----------------------------Add Rating-----------------------------------#
        
        try:
            #Add the rating to rating lsit
            esrb_rating.append(i.find('div', attrs={'class':'content'}).find('img')['src'].replace('.svg', '').split('/')[8].replace('E10plus', 'ET'))
        except:
            esrb_rating.append(np.nan)
            
        #-----------------------------Add content----------------------------------#
        
        try:
            #Get the contetn of the rating, and clean it
            content = i.find('div', attrs={'class':'content'}).find('table').find_all('td')[1].string.split(',')
            content_clean = {i.replace(' ','') for i in content for x in i}

            #Loop through each content and compare it with ratin features, add 1 if exist
            for each in content_clean:
                if each in rating_feat:
                    rating_vector.insert(rating_feat.index(each), 1)
            
            #Append the rating vector
            rating.append(rating_vector)
            
        except:
            rating.append([0]*31) 
            
        #-----------------------------Save Make df------------------------------------#
    
    page_df = pd.DataFrame({'title': title,
                            'console':console,
                            'alcohol_reference': list(list(zip(*rating))[0]),
                            'animated_blood': list(list(zip(*rating))[1]),
                            'blood': list(list(zip(*rating))[2]),
                            'blood_and_gore': list(list(zip(*rating))[3]),
                            'cartoon_violence': list(list(zip(*rating))[4]),
                            'crude_humor': list(list(zip(*rating))[5]),
                            'drug_reference': list(list(zip(*rating))[6]),
                            'fantasy_violence': list(list(zip(*rating))[7]),
                            'intense_violence': list(list(zip(*rating))[8]),
                            'language': list(list(zip(*rating))[9]),
                            'lyrics': list(list(zip(*rating))[10]),
                            'mature_humor': list(list(zip(*rating))[11]),
                            'mild_blood': list(list(zip(*rating))[12]),
                            'mild_cartoon_violence': list(list(zip(*rating))[13]),
                            'mild_fantasy_violence': list(list(zip(*rating))[14]),               
                            'mild_language': list(list(zip(*rating))[15]),
                            'mild_lyrics': list(list(zip(*rating))[16]),
                            'mild_suggestive_themes': list(list(zip(*rating))[17]),
                            'mild_violence': list(list(zip(*rating))[18]),
                            'no_descriptors': list(list(zip(*rating))[19]),
                            'nudity': list(list(zip(*rating))[20]),
                            'partial_nudity': list(list(zip(*rating))[21]),
                            'sexual_content': list(list(zip(*rating))[22]),
                            'sexual_themes': list(list(zip(*rating))[23]),                      
                            'simulated_gambling': list(list(zip(*rating))[24]),
                            'strong_janguage': list(list(zip(*rating))[25]),
                            'strong_sexual_content': list(list(zip(*rating))[26]),                
                            'suggestive_themes': list(list(zip(*rating))[27]),
                            'use_of_alcohol': list(list(zip(*rating))[28]),
                            'use_of_drugs_and_alcohol': list(list(zip(*rating))[29]),
                            'violence': list(list(zip(*rating))[30]),
                            'esrb_rating': esrb_rating
                            }
                          )  

        #----------------------------Write to CSV-----------------------------------#
    #Write each page
    page_df.to_csv('./data/webpages/page{}.csv'.format(pg), index=False)
    
    #Write all of the pages
    page_df.to_csv('./data/webpages/all.csv', mode='a' ,index=False, header=False)
    
    #Tracking pointer
    print(f'Page ===> [{pg}] ===> successfully scraped!')
```

## Duplicated values

- Let's check if there are any duplicated games, due to any error while scraping or duplicated by the website.
- I'll use .value_counts() on title column becuase it must be a unique column.

```python
df.title.value_counts()
```

- As I can see we have some duplicated games, and it's not a lot.
- We will drop it and keep the last occurrence of it.

```python
#Drop the dup rows and keep last
df = df.drop_duplicates(subset='title', keep='last')
```

```python
#Let's check the if any still there.
df.title.value_counts()
```

- We're good, no duplicated rows.
- Let's check if there are any row mislabeled.

```python
#Print each column .values_count()
for i in df.columns:
    print('='*50,'\n')
    print(df[i].value_counts())
```

Good, every column contains both 0's and 1's, which is mean there are no errors while scraping the vector of features.
